```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(precision=3)
```

```python
!gdown --id '1H4BoLWR9xf3oNCAeM1t1y_TeWTwznUfF' --output dataset.csv
dataset = pd.read_csv('dataset.csv')
dataset.describe()
```

```python
# zero-center dataset

dataset = np.array(dataset)

def center_dataset(dataset):

    norm_dataset = np.mean(dataset)

    return norm_dataset

norm_dataset = center_dataset(dataset)
```

```python
# Principal Component Analysis: reduce to 3 dimensions

# Calculate covariance matrix
n = dataset.shape[0]
cov_dataset = (dataset-norm_dataset)@(dataset-norm_dataset).T/n

# Calculate eigenvalues and eigenvectors of covariance matrix
eig_val, eig_vec = np.linalg.eig(cov_dataset)

# Sort eigenvalues and eigenvectors based on absolute eigenvalues
idx = np.argsort(np.abs(eig_val))[::-1]
eig_val = eig_val[idx][:3]
eig_vec = eig_vec[:, idx][:, :3]

print('eigenvalues:', eig_val)
print('eigenvectors:', eig_vec)
```

```python
# Calculate proportion of variance explained

variance = eig_val / np.sum(eig_val)

print('proportion of variance explained:', variance)
```
